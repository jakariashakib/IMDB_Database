## Changing the Scale of MetaCritic 

```python
df['MetaCritic'] = df['MetaCritic']/10
```

## Actor Total Likes 

```python
actor_likes = df.pivot_table(values=['actor_1_facebook_likes', 'actor_2_facebook_likes', 'actor_3_facebook_likes'],
              aggfunc='sum', index=['actor_1_name', 'actor_2_name', 'actor_3_name'])
actor_likes
```

```python
actor_likes['total_likes'] = actor_likes['actor_1_facebook_likes'] + actor_likes['actor_2_facebook_likes'] + actor_likes['actor_3_facebook_likes']
actor_likes['total_likes']
```

```python
# sorting using total_likes and using the same variable (inplace=True)
actor_likes.sort_values(by='total_likes', ascending=False, inplace=True)
actor_likes
```

```python
actor_likes.reset_index(inplace=True)
actor_likes=actor_likes.iloc[:5, :]
```

## Find the most Popular Trios

```python
j = 0
for i in actor_likes['total_likes']:
    temp = sorted([actor_likes.loc[j,'actor_1_facebook_likes'], actor_likes.loc[j,'actor_2_facebook_likes'], actor_likes.loc[j,'actor_3_facebook_likes']])
    if temp[0]>=temp[1]/2 and temp[0]>=temp[2]/2 and temp[1]>=temp[2]/2:
        print(sorted([actor_likes.loc[j,'actor_1_name'], actor_likes.loc[j,'actor_2_name'], actor_likes.loc[j,'actor_3_name']]))
        
    j = j+1
```

# Combine the Dataframe by Genres

```python
df_by_genre = df.loc[:,'CVotes10':'VotesnUS']
df_by_genre.head(5)
```

```python
df_by_genre[['genre_1','genre_2','genre_3']] = df[['genre_1','genre_2','genre_3']]
df_by_genre.head(5)
```

## Create a column cnt and initialize it to 1

```python
df_by_genre['cnt']=1
```

## Group the dataframe by individual genres

```python
df_by_genre[['genre_1', 'genre_2', 'genre_3']]
```

```python
df_by_g1 = df_by_genre.groupby('genre_1').aggregate(np.sum)
df_by_g2 = df_by_genre.groupby('genre_2').aggregate(np.sum)
df_by_g3 = df_by_genre.groupby('genre_3').aggregate(np.sum)
```

### df_by_g1 to df_by_g3

```python
df_by_g1.head(3)
```

```python
df_by_g2.head(3)
```

```python
df_by_g3.head(3)
```

## Add the grouped dataframe and store it in a new dataframe

```python
df_add = df_by_g1.add(df_by_g2)
df_add=df_add.add(df_by_g3)
df_add
```

## Add the grouped dataframe and store it in a new dataframe and remove the null values

```python
df_add = df_by_g1.add(df_by_g2, fill_value=0)
df_add=df_add.add(df_by_g3, fill_value=0)
df_add
```

## Extract genres with atleast 10 occurences

```python
genre_top_10 = df_add.loc[df_add['cnt']>10]
genre_top_10 
```

```python
#genre_top_10.sort_values(by='cnt', ascending=False, inplace=True)
#genre_top_10
```

## Take the mean for every column by deviding with cnt

```python
genre_top_10.iloc[:, 0:-1]= genre_top_10.iloc[:, 0:-1].divide(genre_top_10['cnt'], axis=0)
genre_top_10
# no need cnt
```

## round off the columns of votes to two decimals

```python
genre_top_10.loc[:,'VotesM':'VotesnUS']=round(genre_top_10.loc[:,'VotesM':'VotesnUS'], 2)
genre_top_10
```

## converting CVotes to int type

```python
genre_top_10[genre_top_10.loc[:,'CVotes10':'CVotesnUS'].columns]=genre_top_10[genre_top_10.loc[:,'CVotes10':'CVotesnUS'].columns].astype(int)
genre_top_10
```

## countplot for genres

```python
plt.figure(figsize=(20,8))

ax = sns.barplot(x=genre_top_10['cnt'], y=genre_top_10.index)
# genre top 10 is index not columns
ax.bar_label(ax.containers[0], padding=10, fmt='%.2f',)
plt.show()
```
